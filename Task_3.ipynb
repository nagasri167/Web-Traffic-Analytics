```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = {
    "User_ID": [1, 2, 3, 4, 5, 6, 7, 8],
    "Page_Visited": [
        "Home", "Products", "Pricing", "Blog",
        "Products", "Contact", "Pricing", "Home"
    ],
    "Time_Spent_Min": [2, 5, 4, 3, 6, 1, 5, 2],
    "Bounce": [0, 0, 0, 1, 0, 1, 0, 0]
}

df = pd.DataFrame(data)
```

```python
print("Website Traffic Dataset")
print(df)
```

```python
page_views = df["Page_Visited"].value_counts()

print("\nPage Views")
print(page_views)
```

```python
avg_time = df.groupby("Page_Visited")["Time_Spent_Min"].mean()

print("\nAverage Time Spent on Pages")
print(avg_time)
```

```python

bounce_rate = (df["Bounce"].sum() / len(df)) * 100

print("\nBounce Rate: ", bounce_rate, "%")
```

```python
plt.figure(figsize=(8,5))
page_views.plot(kind='bar')

plt.title("Most Viewed Pages")
plt.xlabel("Pages")
plt.ylabel("Views")

plt.show()
```

```python
plt.figure(figsize=(8,5))
avg_time.plot(kind='bar')

plt.title("Average Time Spent on Pages")
plt.xlabel("Pages")
plt.ylabel("Time (Minutes)")

plt.show()
```
